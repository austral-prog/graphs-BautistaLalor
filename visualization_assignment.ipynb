```python
import matplotlib.pyplot as plt
import numpy as np
```

### 1. Grafico Circular: Distribución de telefonos en China
Este gráfico muestra una visualización de la distribución del mercado de smartphones en China. Un gráfico circular fue elegido porque es la forma más intuitiva de mostrar como el 100% del mercado está dividido por partes.

```python
# Datos simulados de cuota de mercado para China (porcentajes)
marcas = ['Huawei', 'Apple', 'Vivo', 'Honor', 'Xiaomi', 'Oppo', 'Otros']
cuota_mercado = [22.5, 18.2, 16.5, 15.3, 13.8, 10.5, 3.2]

# Colores y separacion (explode) para enfatizar la primera marca
colores = ['#ff6666', '#c2c2d6', '#66b3ff', '#99ff99', '#ffcc99', '#ffb3e6', '#c2f0c2']
separacion = (0.1, 0, 0, 0, 0, 0, 0)

# Crear el grafico
plt.figure(figsize=(8, 8))
plt.pie(cuota_mercado, explode=separacion, labels=marcas, colors=colores, 
        autopct='%1.1f%%', shadow=True, startangle=140, 
        textprops={'fontsize': 11, 'fontweight': 'bold'})

# Agregar titulo
plt.title('Distribución del Mercado de Smartphones en China', fontsize=16, fontweight='bold', pad=20)

# Exportar y mostrar (con manejo de errores si no existe la carpeta 'imgs')
try:
    plt.savefig('imgs/distribucion_mercado_pie.png', dpi=300, bbox_inches='tight')
except FileNotFoundError:
    print("Aviso: Carpeta 'imgs' no encontrada. Guardando en el directorio actual.")
    plt.savefig('distribucion_mercado_pie.png', dpi=300, bbox_inches='tight')

plt.show()
```

### 2. Grafico de Barras: Consumo de Agua vs. Refrescos en America Latina
Esta visualización compara el consumo anual estimado per cápita de agua embotellada y refrescos (sodas) en cuatro de los principales países de América Latina. Un gráfico de barras agrupadas es ideal aquí porque nos permite comparar dos categorías diferentes Agua y Refresco lado a lado dentro de grupos distintos Países.

```python
# Datos simulados de consumo (Litros per capita anualmente)
paises = ['México', 'Brasil', 'Argentina', 'Colombia']
consumo_agua = np.array([250, 180, 160, 110])
consumo_refresco = np.array([150, 115, 130, 85])

x = np.arange(len(paises))  # Posiciones de las etiquetas
ancho = 0.35  # Ancho de las barras

# Crear el grafico
fig, ax = plt.subplots(figsize=(10, 6))
barras1 = ax.bar(x - ancho/2, consumo_agua, ancho, label='Agua Embotellada', color='#4a90e2', edgecolor='black')
barras2 = ax.bar(x + ancho/2, consumo_refresco, ancho, label='Refresco', color='#d0021b', edgecolor='black')

# Agregar etiquetas, título y personalizar el eje X
ax.set_ylabel('Consumo Anual (Litros per Cápita)', fontsize=12)
ax.set_title('Consumo de Agua vs. Refrescos en América Latina', fontsize=15, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(paises, fontsize=11)
ax.legend(loc='upper right', framealpha=1)

# Exportar y mostrar
try:
    plt.savefig('imgs/consumo_bebidas_barras.png', dpi=300, bbox_inches='tight')
except FileNotFoundError:
    print("Aviso: Carpeta 'imgs' no encontrada. Guardando en el directorio actual.")
    plt.savefig('consumo_bebidas_barras.png', dpi=300, bbox_inches='tight')

plt.show()
```

### 3. Grafico de Lineas: Videos Acumulados en YouTube a lo Largo del Tiempo
Este gráfico rastrea una simulación del crecimiento exponencial del número total de videos subidos a YouTube desde su creación. Un gráfico de líneas es la opción estándar y más efectiva para datos de series temporales, ya que los puntos conectados ilustran claramente la trayectoria, la velocidad de crecimiento y la tendencia general.

```python
# Generar datos simulados de series temporales exponenciales
anios = np.arange(2005, 2025)
# Simulando un crecimiento exponencial rápido en miles de millones
videos_billones = 0.01 * np.exp(0.35 * (anios - 2005)) 

# Crear el gráfico
plt.figure(figsize=(11, 6))
plt.plot(anios, videos_billones, color='#FF0000', linewidth=3, marker='o', markersize=6, markerfacecolor='white')

# Rellenar el área bajo la curva por estética
plt.fill_between(anios, videos_billones, color='#FF0000', alpha=0.1)

# Añadir etiquetas y título
plt.title('Videos Acumulados en YouTube (Simulación 2005 - 2024)', fontsize=15, fontweight='bold')
plt.xlabel('Año', fontsize=12)
plt.ylabel('Total de Videos (Miles de Millones)', fontsize=12)

# Asegurar que el eje X muestre solo años enteros
plt.xticks(np.arange(2005, 2025, 2))
plt.grid(True, linestyle='--', alpha=0.7)

# Exportar y mostrar
try:
    plt.savefig('imgs/crecimiento_youtube_lineas.png', dpi=300, bbox_inches='tight')
except FileNotFoundError:
    print("Aviso: Carpeta 'imgs' no encontrada. Guardando en el directorio actual.")
    plt.savefig('crecimiento_youtube_lineas.png', dpi=300, bbox_inches='tight')

plt.show()
```

### 4. Grafico de Dispersion: Rendimiento de Chips (Ley de Moore)
Esta visualización traza el recuento de transistores de varios microprocesadores lanzados entre 1970 y 2020, ilustrando el principio de la Ley de Moore. Un gráfico de dispersión con un eje Y logarítmico es la visualización perfecta para revelar cómo estos puntos de datos históricos independientes forman una correlación lineal cuando hay un crecimiento exponencial.

```python
# Generar datos simulados que representan la Ley de Moore
np.random.seed(42)
anios_lanzamiento = np.random.uniform(1970, 2024, 60)
# Ecuacion base: el recuento se duplica aprox. cada 2 años
recuento_transistores = 2000 * (2 ** ((anios_lanzamiento - 1970) / 2)) 
# Añadir ruido aleatorio para simular la varianza del mundo real
recuento_transistores *= np.random.uniform(0.5, 1.5, 60)

# Crear el gráfico
plt.figure(figsize=(10, 6))
plt.scatter(anios_lanzamiento, recuento_transistores, color='#2c3e50', alpha=0.7, edgecolors='white', s=70)

# Agregar etiquetas, titulo y establecer el eje Y a escala logarítmica
plt.yscale('log')
plt.title("Transistores en Chips de Computadora (Ley de Moore)", fontsize=15, fontweight='bold')
plt.xlabel('Año de Lanzamiento', fontsize=12)
plt.ylabel('Cantidad de Transistores (Escala Logarítmica)', fontsize=12)

# Exportar y mostrar
try:
    plt.savefig('imgs/rendimiento_chips_dispersion.png', dpi=300, bbox_inches='tight')
except FileNotFoundError:
    print("Aviso: Carpeta 'imgs' no encontrada. Guardando en el directorio actual.")
    plt.savefig('rendimiento_chips_dispersion.png', dpi=300, bbox_inches='tight')

plt.show()
```

### 5. Grafico Personalizado: Cuota de Mercado de SO (Graficos de Donas)
Esta figura personalizada utiliza subgráficos (subplots) para colocar dos gráficos circulares modificados —conocidos como "gráficos de dona"— uno al lado del otro. Se eligió este enfoque personalizado para comparar simultáneamente la distribución mundial de la cuota de mercado de los Sistemas Operativos en dos plataformas diferentes (Escritorio vs. Móvil) en un formato visual limpio y moderno.

```python
# Datos Simulados
so_escritorio = ['Windows', 'macOS', 'Linux', 'Chrome OS']
cuota_escritorio = [72.5, 15.3, 3.5, 8.7]

so_movil = ['Android', 'iOS', 'Otros']
cuota_movil = [70.5, 28.5, 1.0]

# Colores
colores_escritorio = ['#0078D7', '#A2AAAD', '#F5A900', '#4285F4']
colores_movil = ['#3DDC84', '#000000', '#808080']

# Crear los subgráficos
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
fig.suptitle('Cuota de Mercado Mundial de Sistemas Operativos', fontsize=16, fontweight='bold')

# Subgráfico 1: Dona de SO de Escritorio
wedges1, texts1, autotexts1 = ax1.pie(cuota_escritorio, labels=so_escritorio, autopct='%1.1f%%', 
                                      startangle=90, colors=colores_escritorio, 
                                      wedgeprops=dict(width=0.4, edgecolor='w'))
ax1.set_title('Plataformas de Escritorio', fontsize=13)

# Subgráfico 2: Dona de SO Móvil
wedges2, texts2, autotexts2 = ax2.pie(cuota_movil, labels=so_movil, autopct='%1.1f%%', 
                                      startangle=90, colors=colores_movil, 
                                      wedgeprops=dict(width=0.4, edgecolor='w'))
ax2.set_title('Plataformas Móviles', fontsize=13)

# Estilizar el texto de los porcentajes para mejor contraste
for autotext in autotexts1 + autotexts2:
    autotext.set_color('white')
    autotext.set_weight('bold')
    
# Cambiar textos específicos a negro donde el fondo es muy claro para leer blanco
autotexts1[1].set_color('black') # macOS
autotexts1[3].set_color('black') # ChromeOS

plt.tight_layout()

# Exportar y mostrar
try:
    plt.savefig('imgs/cuota_mercado_so_donas.png', dpi=300, bbox_inches='tight')
except FileNotFoundError:
    print("Aviso: Carpeta 'imgs' no encontrada. Guardando en el directorio actual.")
    plt.savefig('cuota_mercado_so_donas.png', dpi=300, bbox_inches='tight')

plt.show()
```

## Fuentes de Datos y Atribución

Para la elaboración de las visualizaciones en este cuaderno, los datos fueron introducidos directamente mediante arrays de NumPy, pero están basados en métricas y estimaciones de la industria del mundo real. A continuación se detallan las fuentes concretas de donde se extrajo (o estimó) la información para cada gráfico:

* **1. Mercado de Smartphones en China:** Datos de cuota de mercado basados en los reportes trimestrales del mercado de telefonía móvil en China publicados por **Counterpoint Research** y **Canalys** (estimaciones del Q4 2023 / Q1 2024). https://www.counterpointresearch.com/ (Publican regularmente el "Market Monitor" de smartphones). https://www.canalys.com/ (Tienen una sección de "Newsroom" con los reportes de envíos de celulares).
* **2. Consumo de Bebidas en América Latina:** Volúmenes de consumo anual per cápita basados en los informes de la industria de bebidas no alcohólicas de **Statista** y estudios de mercado de **Euromonitor International** para la región latinoamericana. https://www.statista.com/ (Portal general de estadísticas, muy usado para consumo de bebidas). https://www.euromonitor.com/ (Especialistas en investigación de mercado de bienes de consumo).
* **3. Crecimiento de YouTube:** La curva de crecimiento de videos acumulados se modeló tomando como referencia los hitos históricos de subida de horas de video por minuto publicados en el **Blog Oficial de YouTube** y estimaciones históricas del portal **Tubefilter**. Statista: https://www.statista.com/ (Portal general de estadísticas, muy usado para consumo de bebidas). https://blog.youtube/ (Donde YouTube anuncia sus hitos oficiales). https://www.tubefilter.com/ (Medio especializado en la economía de los creadores y métricas de video).
* **4. Ley de Moore (Chips de Computadora):** El recuento de transistores y las fechas de lanzamiento de los microprocesadores utilizan como referencia el conjunto de datos "Microprocessor Trend Data" recopilado originalmente por Karl Rupp y alojado en **Our World in Data**. https://ourworldindata.org/technological-change (Acá está el dataset exacto sobre la Ley de Moore y los microprocesadores).
* **5. Cuota de Mercado de Sistemas Operativos:** Los porcentajes globales para escritorio y dispositivos móviles fueron extraídos de las estadísticas en tiempo real proporcionadas por **StatCounter Global Stats** (promedios globales recientes). https://gs.statcounter.com/ (Herramienta interactiva para ver la cuota de mercado de sistemas operativos en tiempo real).
